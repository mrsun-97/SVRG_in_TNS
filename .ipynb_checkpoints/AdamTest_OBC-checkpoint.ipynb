```julia
include("BaseModMPS_OBC.jl")
using Plots

global D=6
global spin_deg=2
global mps_size=12
global Tree = generate_BTree(mps_size)

H = generate_Heisenberg_Hamiltonian(-1,-1,-1)
E0 = -0.25*(mps_size-1)
# H = generate_Heisenberg_Hamiltonian(1,1,1)
# E0 = -0.75*mps_size+sin(2π/mps_size)/(1-cos(2π/mps_size))

spH = sparse_indices(H)
mps0 = generate_MPS(rand; D=D, spin_deg=spin_deg, mps_size=mps_size)
# print_mat(mps.site[1].data)

println("exact energy = ", exact_energy!(mps0))

println("")
# print_mat(mps.site[1].data)
println("E0 = ",E0)
```

#### SGD 预处理

```julia
step = 1e-2
data1s, mps = mkSGD(mps0, H; step=10,chainlen=1000,η=step, progress=true)
println("done")
dt = log10.(abs.((data1s.-E0)))
plot(dt, title="preSGD L=$mps_size D=$D Markov_len=$mk_len",color=:coral, size=[900,500], lw=2,
            xlabel="steps",ylabel="E", label="step=$step")
```

### SVRG

```julia
svrg_step = 3e-3
data1v, mps3 = mkReweightSVRG(mps, H; outer=80,inner=10,long=2500,short=500,η=svrg_step)
println("done")
```

### Tests

```julia
using Plots.PlotMeasures
# auto-test
# len  = [400,300,200,100]
# step = [0.001,0.005,0.02,0.04]
Nstep = 10*60

# mps  = mps1
len  = [500,2000]
step = [0.002, 0.006, 0.02]

data = Array{Vector{Float64},3}(undef,3,length(len),length(step))
set  = [(i,j) for i in 1:length(len) for j in 1:length(step)]

global progress1 = Progress(length(len)*length(step))
for (i, j) in set
    data2a, _ = mkAdam(mps, H; step=Nstep, chainlen=len[i], η=step[j])
    data2s, _ =  mkSGD(mps, H; step=Nstep, chainlen=len[i], η=step[j])
    data2v, _ = mkReweightSVRG(mps, H; outer=Int(Nstep/10),inner=10,long=Int(len[i]/2*8),short=Int(len[i]/2),η=step[j])
    data[:,i,j] .= data2a, data2v, data2s
    next!(progress1)
end

println("done")
```

```julia
gr(legend=:bottomleft)
figure = Array{Plots.Plot{Plots.GRBackend},2}(undef,length(len),length(step))
f(x) = log10.(abs.(x .- E0))
# f(x) = log10.(x .- E0)
# f(x) = x .- E0
for i in 1:length(len)
    for j in 1:length(step)
#         ld = map(f,data[:,i,j])
        ld = data[:,i,j]
        tmp = plot(f.(ld), title="len=$(len[i])  step=$(step[j])",labels=["Adam" "SVRG" "SGD"])
        figure[i,j] = tmp
    end
end
tf = plot(figure..., ms=[3.5,4.5,3.5],size=[1200,800], markerstrokewidth=0, xlabel="step", ylabel="Log(E)",
        ylims=(-15,-1),layout=(length(step),length(len)) )
```
