```julia
include("BaseModMPS_PBC.jl")
using Plots

global D=6
global spin_deg=2
global mps_size=16
global Tree = generate_BTree(mps_size)

H = generate_Heisenberg_Hamiltonian(-1,-1,-1)
E0 = -0.25*(mps_size)
# H = generate_Heisenberg_Hamiltonian(1,1,1)
# E0 = (0.25-0.696393522538549)*mps_size  # for N = 16

spH = sparse_indices(H)
mps0 = generate_MPS(rand; D=D, spin_deg=spin_deg, mps_size=mps_size)
# print_mat(mps.site[1].data)

println("appr. energy = ",exact_energy!(mps0))

println("")
# print_mat(mps.site[1].data)
println("E0 = ",E0)
```

#### SGD 预处理

```julia
step = 1e-2
mk_len=1000
data1s, mps = mkSGD(mps0, H; step=40,chainlen=mk_len,η=step, progress=true)
_dt = log10.(abs.((data1s.-E0)))
scatter(_dt, title="preSGD L=$mps_size D=$D Markov_len=$mk_len", ms=2.2,color=:coral, size=[900,500],markerstrokewidth=0,
            xlabel="steps",ylabel="Log(E)", label="L=$mps_size, D=$D")
```

### Tests

```julia
using Plots.PlotMeasures
# auto-test
# len  = [400,300,200,100]
# step = [0.001,0.005,0.02,0.04]
Nstep = 10*80

# mps  = mps1
len  = [500, 2000]
step = [0.001, 0.04]

data = Array{Vector{Float64},3}(undef,3,length(len),length(step))
set  = [(i,j) for i in 1:length(len) for j in 1:length(step)]

global progress1 = Progress(length(len)*length(step))
Threads.@threads for (i, j) in set
    data2a, _ = mkAdam(mps, H; step=Nstep, chainlen=len[i], η=step[j], progress=false)
    data2s, _ =  mkSGD(mps, H; step=Nstep, chainlen=len[i], η=step[j], progress=false)
    data2v, _ = mkReweightSVRG(mps, H; outer=Int(Nstep/10),inner=10,long=Int(len[i]/2*8),short=Int(len[i]/2),η=step[j], progress=false)
    data[:,i,j] .= data2a, data2v, data2s
    next!(progress1)
end

println("done")
```

```julia
gr(legend=:bottomleft)
figure = Array{Plots.Plot{Plots.GRBackend},2}(undef,length(len),length(step))
f(x) = log10.(abs.(x .- E0))
# f(x) = x .- E0
for i in 1:length(len)
    for j in 1:length(step)
#         ld = map(f,data[:,i,j])
        ld = data[:,i,j]
        tmp = scatter(f.(ld), title="len=$(len[i])  step=$(step[j])",labels=["Adam" "SVRG" "SGD"])
        figure[i,j] = tmp
    end
end
tf = plot(figure..., ms=2, size=[1200,800], markerstrokewidth=0, xlabel="step", ylabel="Log(E)",
        ylims=(-8,-1),
#         ylims=(-0.001,0.001),
        layout=(length(step),length(len)) )
```
