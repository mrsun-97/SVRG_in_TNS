```julia
include("BaseModMPS_PBC.jl")
using Plots
Threads.nthreads()
```

```julia

idn = [1.0 0; 0 1.0]
σx = [0  1;1  0]
iσy = [0 1;-1  0]
σz = [1  0;0  -1]

H2 = +(
     kron(σx,  σx)/4,
    -kron(iσy,iσy)/4,
     kron(σz,  σz)/4
    )

show(stdout, MIME"text/plain"(), H2)
```

### 2-site OBC
##### O---O

```julia
# N=2
eigen(H2)
```

### 3-site PBC
##### --O---O---O--

```julia
# N=3
H3 = +(
    kron(H2, idn),
    kron(idn, H2),
    +(
        kron(σx, idn, σx)/4,
       -kron(iσy,idn,iσy)/4,
        kron(σz, idn, σz)/4
    )
)
h = eigen(H3)
```

```julia
# export as LaTeX codes
for i in 1:8
    for j in 1:7
        print(round(h[i,j],digits=3)," & ")
    end
    print(round(h[i,8],digits=3),"\\\\\n")
end
```

### 4-site PBC
##### --O---O---O---O--

```julia
# N = 4, PBC
H4 = +(
    kron(H2, idn, idn),
    kron(idn, H2, idn),
    kron(idn, idn, H2),
    +(
        kron(σx, idn, idn, σx)/4,
       -kron(iσy,idn, idn,iσy)/4,
        kron(σz, idn, idn, σz)/4
    )
)
```

```julia
vals, vecs = eigen( Symmetric(H4) )
```

## Calculation starts here.

```julia
include("BaseModMPS_PBC.jl")

global D=3
global spin_deg=2
global mps_size=2
global Tree = generate_BTree(mps_size)
arr = [0.0, -1.5, -0.75, -2.0]

H = generate_Heisenberg_Hamiltonian(1,1,1)
# E0 = (0.25-0.696393522538549)*mps_size  # for N = 16
E0 = arr[mps_size]

spH = sparse_indices(H)
mps0 = generate_MPS(rand; D=D, spin_deg=spin_deg, mps_size=mps_size)
# print_mat(mps.site[1].data)

println("appr. energy = ",exact_energy!(mps0))

println("")
# print_mat(mps.site[1].data)
println("E0 = ",E0)
```

### SGD

```julia
step = 1e-3
mk_len = 5000
data1s, mps1 = mkSGD(mps0, H; step=8000,chainlen=mk_len,η=step, progress=true)
println("done")
```

```julia
dt = log10.(abs.((data1s.-E0)))
# dt = data1s.-E0
figs = scatter(dt, title="SGD, step=$step, Markov_len=$mk_len", ms=2.2, color=:coral, markerstrokewidth=0,
            xlabel="steps", ylabel="Log(E)", label="L=$mps_size, D=$D", size=[900,500])
savefig(figs,"./heisenberg_tex/SGD_2site_PBC.pdf")
figs
```

### Adam

```julia
η = 1.0e-3
data1a, mps2 = mkAdam(mps0, H; step=2000,chainlen=8000,η=η, progress=true)
println("done")
```

```julia
dt = log10.(abs.(data1a.-E0))
# dt = data1a .- E0
figa = scatter(dt, title="Adam, step=$η",ms=2.2, color=:red, size=[1000,500],markerstrokewidth=0,
            xlabel="steps",ylabel="Log(E)", label="L=$mps_size, D=$D")
savefig(figa, "heisenberg_tex/Adam_2site_PBC.pdf")
figa
```

```julia
data2a, mps22 = mkAdam(mps2, H; step=500,chainlen=10000,η=η, progress=true);
```

```julia
dt = log10.(abs.(data2a.-E0))
figa = scatter(dt, title="Adam, step=$η",ms=2.2, color=:red, size=[1000,500],markerstrokewidth=0,
            xlabel="steps",ylabel="Log(E)", label="L=$mps_size, D=$D")
```

```julia
svrg_step = 1.0e-3
outer = 3000
inner = 5
data1v, mps3 = mkReweightSVRG(mps0, H; outer=outer,inner=inner,long=5000,short=1000,η=svrg_step, progress=true)
println("done")
```
